# CALCULATING OUTLIERS

```python
def Remove_outlier(new_df):  
    Q1 = new_df.quantile(0.25)
    Q3 = new_df.quantile(0.75)
    IQR = Q3 - Q1

    new_df = new_df[~((new_df < (Q1 - 1.5 * IQR)) |(new_df > (Q3 + 1.5 * IQR))).any(axis=1)]

    return new_df
    # plt.show()

df2 = df.copy()

print(df2.shape)
```

# Feature Engineering

```python
y.value_counts()
```

We notice that their is a class imbalance in the quality. We could apply a technique to rectify it

In type column 6 is the majority class and it makes sense to do oversampling on the minority classes that is duplicating random records from the minority classes. One limitation to this method is that it could cause overfitting as there is a risk of high bias.

SMOTE (Synthetic Minority Oversampling Technique) works by randomly picking a point from the minority class and computing the k-nearest neighbors for this point. The synthetic points are added between the chosen point and its neighbors.

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4)
```

```python
#Transforming the data after SMOTE
x,y = oversample.fit_resample(x,y)
```

# PREDICTION

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x.values,y.values,test_size=0.25,random_state=21)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
                                        
```

Feature Scaling:-
Feature scaling is important after applying over/under sampling methods as it would take the accuracy higher

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# SUPERVISED METHODS

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {}
# Logistic Regression
lr = LogisticRegression()
models["Logistic Regression"] = lr
# Decision Tree Classifier
dtc = DecisionTreeClassifier()
models["Decision Tree Classifier"] = dtc
rf = RandomForestClassifier(n_estimators=10, criterion="entropy",random_state=0)
models["Random Forest"] = rf
```

```python
models
```

```python
for model in models:
    models[model].fit(x_train,y_train)
```

```python
from sklearn.metrics import multilabel_confusion_matrix
```

```python
def model_performance(modelName,model,x_test,y_test):
    print("_______________________________________________")    
    print("Model:",modelName)
    y_pred = model.predict(x_test)
    print("Accuracy Score:",accuracy_score(y_test,y_pred))
    cm = multilabel_confusion_matrix(y_test,y_pred)
    print("Confusion Matrix:\n",cm)
    print("Classification Report:\n",classification_report(y_test,y_pred))
    plt.show()
    print("_______________________________________________")
```

```python
#Performance
for model in models:
    model_performance(model,models[model],x_test,y_test)
```

Random Forest is the best model with 85% test accuracy

```python
from numpy.linalg import eigh

# covariance matrix

cov_matrix = np.cov(X_kscaled, rowvar=False)

# Calculate eigenvalues and eigenvectors

egnvalues, egnvectors = eigh(cov_matrix)

# Calculating explained variance

total_egnvalues = sum(egnvalues)
var_exp = [(i/total_egnvalues) for i in sorted(egnvalues, reverse=True)]

# Plotting explained variance against cumulative explained variance

import matplotlib.pyplot as plt
cum_sum_exp = np.cumsum(var_exp)
plt.bar(range(0,len(var_exp)), var_exp, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_exp)), cum_sum_exp, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

# Kmeans without PCA

```python
from sklearn.metrics import f1_score
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
#elbow method
#by elbow method we will try to find the most optimal number of k using elbow plot
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=12, random_state=0)
    kmeans.fit(X_kscaled)
    wcss.append(kmeans.inertia_)
f3, ax = plt.subplots(figsize=(8, 6))
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The "K" value of 2 will be selected since the elbow in the graph above can be seen to dip around the number 2.



The dataset will first be clustered using K-Means without using principal component analysis (PCA). Keep in mind that the dataset has 11 total dimensions.

```python
#Applying kmeans to the dataset, set k=2
kmeans = KMeans(n_clusters = 2)
clusters = kmeans.fit_predict(X_kscaled)
labels = kmeans.labels_
```

# Model Evaluation K-means

Sillhoutte score


To assess how thick and clearly divided the clusters are, the silhouette score for a sample set of data points is employed. The silhouette score falls within the range [-1, 1]. High sillhoutte score indicated the data points are matched to their own cluster

```python
#evaluate model
from sklearn import metrics
from sklearn.metrics import pairwise_distances
metrics.silhouette_score(X_kscaled, labels, metric='euclidean')
```

Comparing the best sillhouette score for clusters 2,3,4

```python
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8))
for i in [2, 3, 4]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)
    '''
    Creating SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X_kscaled)
```

Inertia score-



The K-Means clustering performance of a dataset is measured by its inertia. It is determined by calculating the separation between each data point and its centroid, squaring this separation, and adding these squares for each cluster.
